```python
model_name = "hsekhalilian/Speech_To_Text_Finetuning_01"
asr_model = EncDecHybridRNNTCTCBPEModel.from_pretrained(model_name)
```

```python
manifest_path = "/home/jovyan/.cache/asr-persian-leaderboard/datasets/hsekhalilian___fleurs/manifests/validation_manifest.json"
manifest_path = "/home/jovyan/.cache/asr-persian-leaderboard/datasets/hsekhalilian___sorted_commonvoice/manifests/dev_manifest.json"

with open(manifest_path, "r", encoding="utf-8") as f:
    references = []
    predictions = []
    audio_files = []
    for line in f:
        entry = json.loads(line)
        audio_files.append(entry["audio_filepath"].replace("user", "jovyan"))
        references.append(entry["text"])
```

```python
results = asr_model.transcribe(audio_files)
predictions = [result.text for result in results]
```

```python
print("wer: ", wer(references, predictions))
print(evaluate_asr(references, predictions))

for reference, prediction in zip(references, predictions):
    print("="*30)
    print(persian_normalizer_no_punc(reference))
    print(persian_normalizer_no_punc(prediction))
```
