```python
import os

manifest_paths = [
    "/home/jovyan/.cache/asr-finetuning/datasets/hsekhalilian___fleurs/manifests/train_manifest.json",
    "/home/jovyan/.cache/asr-finetuning/datasets/hsekhalilian___sorted_commonvoice/manifests/train_manifest.json"
]

output_dir = "/home/jovyan/.cache/asr-finetuning/datasets/combined"
output_path = os.path.join(output_dir, "train_manifest.json")

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Concatenate the manifest files
with open(output_path, 'w') as outfile:
    for path in manifest_paths:
        with open(path, 'r') as infile:
            for line in infile:
                outfile.write(line)

print(f"Combined manifest saved to: {output_path}")
```

```python
manifest_path = "/home/jovyan/.cache/asr-finetuning/datasets/hsekhalilian___fleurs/manifests/validation_manifest.json"
# manifest_path = "/home/jovyan/.cache/asr-finetuning/datasets/hsekhalilian___sorted_commonvoice/manifests/dev_manifest.json"
references = []
predictions = []
audio_files = []

with open(manifest_path, "r", encoding="utf-8") as f:
    for line in f:
        entry = json.loads(line)
        audio_files.append(entry["audio_filepath"].replace("user", "jovyan"))
        references.append(entry["text"])
```

```python
results = asr_model.transcribe(audio_files)
predictions = [result.text for result in results]
```
